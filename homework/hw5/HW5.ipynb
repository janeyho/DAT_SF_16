```python
#import data
import pandas as pd
data = pd.read_csv('breast-cancer-wisconsin.csv', names = ['ID','Clump_Thickness','Uniform_Cell_Size','Uniform_Cell_Shape','Marginal_Adhesion','Epi_Cell_Size','Bare_Nuclei','Bland_Chromatin','Normal_Nucleoli','Mitoses','Class'])
```

```python
#using less data from overloaded class
less_data_benign = data[data['Class'] == 2]
less_data_benign.info()
```

```python
less_data_benign_samp = less_data_benign.sample(241)
malignant = data[data['Class'] == 4]
```

```python
balanced_data = pd.concat([less_data_benign_samp, malignant])
balanced_data.describe()
```

```python
balanced_data
```

```python
#find question marks and drop
balanced_data[balanced_data.Bare_Nuclei == '?']
```

```python
balanced_data_cleaned = balanced_data.convert_objects(convert_numeric = True).dropna()
```

```python
balanced_data_cleaned.describe()
```

```python
#change 2/benign to 0 and 4/malignant to 1
#balanced_data_cleaned[balanced_data_cleaned.Class == 2]
balanced_data_cleaned.Class.replace([2, 4],[0,1],inplace=True)
```

```python
balanced_data_cleaned.head()
```

```python
df = balanced_data_cleaned
```

```python
target_col = "Class"
features_col =  ['ID','Clump_Thickness','Uniform_Cell_Size','Uniform_Cell_Shape','Marginal_Adhesion','Epi_Cell_Size','Bare_Nuclei','Bland_Chromatin','Normal_Nucleoli','Mitoses']
```

```python
features = df[features_col]
target = df[target_col]
```

```python
target
```

```python
import numpy as np
from sklearn.cross_validation import StratifiedShuffleSplit 
```

```python
X = features.values
y = target.values
```

```python
sss = StratifiedShuffleSplit(y, 3, test_size=0.5, random_state=0)
```

```python
for train_index, test_index in sss:
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
```
